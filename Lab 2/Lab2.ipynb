```python
# Importing the Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Part 1

```python
class _Vehicle:
    def __init__(self, Brand, Model):
        Condition = 'New'
        self.Brand = Brand
        self.Model = Model

    def _Print(self):
        print(f'The Brand is {self.Brand} and the Model is {self.Model}')
```

```python
Object = _Vehicle('Honda', 'Civic')
Object._Print()
```

```python
class Car(_Vehicle):
    def __init__(self, Brand, Model, Type):
        super().__init__(Brand, Model)
        self.Type = Type

    def Print(self):
        print(f'The Brand is {self.Brand}, the Model is {self.Model} and the Type is {self.Type}')
```

```python
Object2 = Car('Suzuki', 'Alto', 'Car')
Object2.Print()
```

```python
Object2._Print()
```

```python
List = ['agoraphobia', 'acrophobia', 'pteromerhanophobia', 'claustrophobia', 'entomophobia']
Count = 0

for String in List:
    if len(String) >= 2 and String[0] == String[-1]:
        Count += 1
print(f'The Number of Strings with the Same First and Last Characters are: {Count}')
```

```python
List = ['Xenon', 'Apple', 'X-Ray', 'Banana', 'Xylophone']
SortedList = sorted(List, key=lambda x: (not x.startswith('X'), x))
print(SortedList)
```

```python
List = [(1, 5), (2, 4), (3, 3), (4, 2), (5, 1)]
List.sort(key=lambda x: x[-1])
print(List)
```

### Part 3

```python
DataFrame = pd.read_csv('Drinks.csv')
print(DataFrame)
```

```python
print('The Number of Rows in the Dataset is:', len(DataFrame))
```

```python
# Dropping NaN Values
DataFrame = DataFrame.dropna()
```

```python
def PrintColumn(DataFrame, Column):
    print(DataFrame[Column])


Column = 'country'
PrintColumn(DataFrame, Column)
```

```python
DataFrame['beer_servings'].plot(kind='hist', ec='black')
plt.title('Frequency of Beer Servings')
plt.xlabel('Beer Servings')
plt.show()
```

```python
plt.scatter(DataFrame['continent'], DataFrame['beer_servings'])
plt.title('Beer Servings per Continent')
plt.xlabel('Continent')
plt.ylabel('Beer Servings')
plt.show()
```

```python
Figure = plt.figure(figsize=(10, 10))
Plot = Figure.add_subplot(projection='3d')
Plot.scatter(DataFrame['beer_servings'], DataFrame['spirit_servings'], DataFrame['wine_servings'])
Plot.set_xlabel('Beer Servings')
Plot.set_ylabel('Spirit Servings')
Plot.set_zlabel('Wine Servings')
plt.show()
```

```python
NewDataFrame = DataFrame.groupby('continent').sum()
Percentages = np.array(NewDataFrame['beer_servings'])
Labels = DataFrame['continent'].unique()
plt.pie(Percentages, labels=Labels, autopct='%1.1f%%')
plt.title('Percentage of Beer Servings per Continent')
plt.show()
```

```python
plt.bar(DataFrame['continent'].unique(), NewDataFrame['beer_servings'], ec='black')
plt.title('Total Beer Servings per Continent')
plt.xlabel('Continent')
plt.ylabel('Total Beer Servings')
plt.show()
```
